```python
import pandas as pd

data = {
    "Name": ["A", "B", "C", "D", "E", "F", "G"],
    "Department": ["IT", "HR", "IT", "Sales", "HR", "IT", "Sales"],
    "Salary": [50000, 40000, 60000, 55000, 45000, 70000, 48000],
    "Age": [25, 28, 30, 35, 26, 32, 29],
    "Experience": [2, 3, 5, 7, 2, 6, 4]
}

df = pd.DataFrame(data)
df
```

```python
df.head(3)
df.shape
df.columns
df[['Name','Salary']]
df['Salary'].mean()
df['Age'].max()
df[df['Salary']>50_000]
df[df['Department']=='IT']
df[df['Age']<30]
df[(df['Salary']>=45_000) & (df['Salary']<=65_000)]
df.query('Department=="HR" and Experience > 2')
df[(df['Age']==25) | (df['Age']==35)]
```

```python
df.assign(Bonus=df['Salary']*0.1)
```

```python
df['Senior']=df['Age'].apply(lambda x: 'True' if x>30 else 'False')
df1=df.copy()
```

```python
df1['Salary']+5000
df
```

```python
df['Tax']=df['Salary'].apply(lambda x:'20%' if x>60000 else '10%')
df
```

```python
df.groupby('Department')['Salary'].mean()
df.groupby('Department')['Salary'].max()
df.groupby('Department').count()
df.groupby('Department')['Experience'].mean()
df.groupby('Department')['Salary'].agg(["min","max"])
```

```python
df.sort_values('Salary',ascending=False)
df.sort_values('Age')
df.sort_values(['Department','Salary'])
```
