```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
df=datasets.load_diabetes(as_frame=True).frame
```

```python
df.head()
```

```python
X=df.drop("target",axis=1)
y=df["target"]
```

```python
X.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,mean_absolute_error

X_train,X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.3,random_state=42)

rfr=RandomForestRegressor(max_depth=4,n_estimators=401,oob_score=True,min_samples_leaf=4,
    random_state=42)
rfr.fit(X_train,y_train)
y_pred=rfr.predict(X_test)
y_train_pred=rfr.predict(X_train)

print("r2 score for training is : ",r2_score(y_train,y_train_pred))
print("r2 score for testing is : ",r2_score(y_test,y_pred))

print("MAE :", mean_absolute_error(y_test, y_pred))
print("MAE :", mean_absolute_error(y_train,y_train_pred))

print(rfr.oob_score_)
```
