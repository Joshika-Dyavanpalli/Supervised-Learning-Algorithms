```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,recall_score,precision_score
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
```

```python
iris_df=pd.read_csv("Iris (1).csv")

X=iris_df.drop("Species",axis=1)
y=iris_df["Species"]

X_train,X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.5,random_state=42)

#LogisticRegression algorithm
pipeline=Pipeline([("scaler",StandardScaler()),("lr_model",LogisticRegression())])
param_grid={"lr_model__C":[0.1,0.2,0.01,1,0.002,2,1.5]}

lr_model=GridSearchCV(
    pipeline,
    param_grid,
    cv=5,
    scoring="recall_macro")

lr_model.fit(X_train,y_train)
y_pred = lr_model.predict(X)

r_score=recall_score(y,y_pred,average="macro")
print(r_score)
a=accuracy_score(y,y_pred)
p=precision_score(y,y_pred,average="macro")

print(lr_model.best_score_) 
print(a)
print(p)
```

```python
#kNN algorithm
iris_df=pd.read_csv("Iris (1).csv")

X=iris_df.drop("Species",axis=1)
y=iris_df["Species"]

X_train,X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.5,random_state=42)

pipeline=Pipeline([("scaler",StandardScaler()),("knn_model",KNeighborsClassifier())])
param_grid={"knn_model__n_neighbors":[3,5,7,9]}

knn_model=GridSearchCV(
    pipeline,
    param_grid,
    cv=5,
    scoring="recall_macro")

knn_model.fit(X_train,y_train)
y_pred = knn_model.predict(X)

r_score=recall_score(y,y_pred,average="macro")
print(r_score)
print(knn_model.best_score_)
a=accuracy_score(y,y_pred)
p=precision_score(y,y_pred,average="macro")
print(a)
print(p)
```

```python
#Naive bayes algorithm-guassiannb
iris_df=pd.read_csv("Iris (1).csv")

X=iris_df.drop("Species",axis=1)
y=iris_df["Species"]

X_train,X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.5,random_state=42)

model=GaussianNB()
model.fit(X_train,y_train)

y_pred=model.predict(X)

r_score=recall_score(y,y_pred,average="macro")
print(r_score)
a=accuracy_score(y,y_pred)
p=precision_score(y,y_pred,average="macro")
print(a)
print(p)
```
